### Handling Missing Values in Age and Embarked

```python
# For Embarked (replcing missing values with mode)
df["Embarked"].fillna("S", inplace=True)
print(df.isnull().sum())
```

```python
# For Age (replcing missing values with mean)
df["Age"].fillna(df["Age"].mean(), inplace=True)
print(df.isnull().sum())
```
